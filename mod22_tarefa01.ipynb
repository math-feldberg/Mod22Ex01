1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).


```python
data.head()
```

```python
df_high = data['High'].stack().reset_index()
df_high.columns = ['data', 'ativo', 'valor']
df_high.head()
```

```python
df_low = data['Low'].stack().reset_index()
df_low.columns = ['data', 'ativo', 'valor']
df_low.head()
```

```python
df_merge = df_high.merge(df_low, left_index=True, right_index=True)
df_merge['amplitude'] = df_merge['valor_x'] - df_merge['valor_y']
df = df_merge.drop(columns = ['valor_x', 'data_y', 'ativo_y', 'valor_y'])
df.columns = ['data', 'ativo', 'valor']
df.head()
```

2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.


```python
fig = px.line(df, x="data", y="valor", color='ativo', template='plotly_white')
fig.show()
```

3. Veja se consegue identificar que ativo apresentou maior volatilidade.
